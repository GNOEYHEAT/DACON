```python
import os
from glob import glob
import numpy as np
import pandas as pd
```

```python
submission_list = glob('/content/drive/MyDrive/DACON-Basic/235892_인구 데이터 기반 소득 예측 경진대회/submission/*.csv')
submission_list
```

```python
temp = []
for _ in glob('/content/drive/MyDrive/DACON-Basic/235892_인구 데이터 기반 소득 예측 경진대회/submission/*.csv'):
    temp.append(pd.read_csv(_)["target"].values)
submissions = np.stack(temp).T

submissions.shape
```

```python
submission=np.where(np.mean(submissions, axis=1)>=0.5, 1, 0)

sample_submission = pd.read_csv("/content/drive/MyDrive/DACON-Basic/235892_인구 데이터 기반 소득 예측 경진대회/data/sample_submission.csv")
sample_submission.target = submission
sample_submission.to_csv("/content/drive/MyDrive/DACON-Basic/235892_인구 데이터 기반 소득 예측 경진대회/submission.csv",index=False)
```
